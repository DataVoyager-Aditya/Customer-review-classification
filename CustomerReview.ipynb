```python
import pandas as pd
import numpy as np
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.preprocessing.text import one_hot
from tensorflow.keras.preprocessing.sequence import pad_sequences
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv(r'C:\Users\Lenovo\Downloads\Restaurant_Reviews.tsv',sep='\t')
```

```python
X= df['Review']
y=df['Liked']
```

```python
vocab_size= 100
X_enc=[]
for i in range(0,1000):
    enc= one_hot(X[i],vocab_size)
    X_enc.append(enc)
X_enc
```

```python
max_padding=30
X= pad_sequences(X_enc,max_padding,padding='post')
X[0]
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2)
```

```python
max_dim= 40
model= keras.Sequential()
model.add(keras.layers.Embedding(vocab_size,max_dim,input_length=max_padding,name='embedding'))
model.add(keras.layers.Flatten())
model.add(keras.layers.Dropout(0.3))
model.add(keras.layers.Dense(100,activation='tanh'))
model.add(keras.layers.Dropout(0.5))
model.add(keras.layers.Dense(50,activation='relu'))
model.add(keras.layers.Dropout(0.8))
model.add(keras.layers.Dense(200,activation='relu'))
model.add(keras.layers.Dense(1,activation='sigmoid'))
```

```python
model.compile(
    optimizer='adam',
    loss= 'binary_crossentropy',
    metrics=['Accuracy']
)
```

```python
model.fit(X_train,y_train,epochs=30)
```

```python
model.evaluate(X_test,y_test)
```
